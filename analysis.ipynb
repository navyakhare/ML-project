```python
for j in range(3):
    colvar=np.loadtxt('ala4-explicit-metad-combine-1ns-10ns-60in-10nsColvar/COLVAR_anc_'+str(j))
    print("Model : "+str(j+1))
    for i in range(3):
        title("phi"+str(i+1)+" vs psi"+str(i+1)+" colored by sumabs")
        plt.xlim(-4,4)
        plt.ylim(-4,4)
        scatter(colvar[:,i*2+1],colvar[:,i*2+2],c=colvar[:,7])
        colorbar()
        show()
```
